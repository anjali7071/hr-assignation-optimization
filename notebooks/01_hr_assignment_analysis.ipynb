```python
import pandas as pd
```

```python
employees = pd.read_csv("../data/employees.csv")
employees
```

```python
tasks = pd.read_csv("../data/tasks.csv")
tasks
```

```python
assignments = []

for _, task in tasks.iterrows():
    # Find employees with matching skill and enough available hours
    eligible_employees = employees[
        (employees["skill"] == task["required_skill"]) &
        (employees["availability_hours"] >= task["required_hours"])
    ]
    
    if not eligible_employees.empty:
        selected_employee = eligible_employees.iloc[0]
        assignments.append({
            "task_id": task["task_id"],
            "task_name": task["task_name"],
            "assigned_to": selected_employee["name"]
        })
    else:
        assignments.append({
            "task_id": task["task_id"],
            "task_name": task["task_name"],
            "assigned_to": "No suitable employee"
        })

assignments
```

```python
assignment_df = pd.DataFrame(assignments)
assignment_df
```

```python
assignment_df.to_csv("../results/task_assignments.csv", index=False)
```
